# Successive Differentiation

```python
from torch.autograd import grad

def nth_derivative(f, wrt, n=2):
    for i in range(n):
        grads = grad(f, wrt, create_graph=True)
        print(grads)
        f = grads[0].sum()
        print(f"f_equation : {f} ")
    
    return grads
```

```python
import torch
x = torch.tensor(5.0, requires_grad = True)
```

```python
f = x**2 + x**3
```

```python
f.grad_fn
```

```python
nth_derivative(f, x)
```
